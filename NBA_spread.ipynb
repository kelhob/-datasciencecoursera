```python
q1= """ select games.*,
                  spread.spread1
                   from nba_df games
                  join nbaspread_df spread on games.game_id = spread.game_id
                  and games.team_id=spread.team_id
              where spread.book_id = 238
              and games.game_date>'1/1/1990'
              """
print(ps.sqldf(q1, locals()))
```

```python
q2= """ select games.*,
                  spread.spread2
                   from nba_df games
                  join nbaspread_df spread on games.game_id = spread.game_id
                  and games.team_id=spread.a_team_id
              where spread.book_id = 238
              and games.game_date>'1/1/1990'
              """
print(ps.sqldf(q2, locals()))
```

```python
# prompt: union q1 and q2

q1= """ select games.*,
                  spread.spread1
           from nba_df games
           inner join nbaspread_df spread
           on games.game_id = spread.game_id
           and games.team_id=spread.team_id
           where spread.book_id = 238
           and games.game_date>'1/1/1990'
           """
q1_df = ps.sqldf(q1, locals())

q2= """ select games.*,
                  spread.spread2
            from nba_df games
            inner join nbaspread_df spread
            on games.game_id = spread.game_id
            and games.team_id=spread.a_team_id
            where spread.book_id = 238
            and games.game_date>'1/1/1990'
            """
q2_df = ps.sqldf(q2, locals())

# Union the two dataframes
union_df = pd.concat([q1_df, q2_df], ignore_index=True)
union_df
```

```python
# prompt: write union_df to a csv file in drive/myDrive/data folder

union_df.to_csv('/content/drive/My Drive/data/union_df.csv', index=False)
```
